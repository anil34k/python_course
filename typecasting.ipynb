types of typecasting in python
1. implicit typecasting
2. explicit typecasting


```python
# implicit typecasting - it occurs when python automatically converts one datatype to another during the execution

a = 7 # int
b = 4 # int
c = a + b # int + int = int
print(type(c)) # <class 'int'>

e = 7.5 # float
f = 4 # int
g = e + f # float + int = float
print(type(g)) # <class 'float'>
```

```python
# explicit typecasting - it is when a programmer will do manual changes the value of the datatype using the built in typecasting functions

# typecasting on interger datatype
# syntax
# variable_name = value
# print(data_type(variable_name))

a = 7.5 # float
b = int(a) # typecasting float to int
print(b) # 7
print(type(b)) # <class 'int'>
```

```python
# interger datatype - can be converted into float, complex, string, boolean

a = 98 # int
print(float(a)) # 98.0
print(complex(a)) # 98+0j
print(str(a)) # '98'
print(bool(a)) # True

print(list(a)) # TypeError: 'int' object is not iterable
print(tuple(a)) # TypeError: 'int' object is not iterable
print(set(a)) # TypeError: 'int' object is not iterable
print(dict(a)) # TypeError: 'int' object is not iterable
```

```python
# float datatype - can be converted into int, complex, string, boolean

a = 7.5 # float
print(int(a)) # 7
print(complex(a)) # 7.5+0j
print(str(a)) # '7.5'   
print(bool(a)) # True

print(list(a)) # TypeError: 'float' object is not iterable
print(tuple(a)) # TypeError: 'float' object is not iterable
print(set(a)) # TypeError: 'float' object is not iterable
print(dict(a)) # TypeError: 'float' object is not iterable
```

```python
# complex datatype - can be converted into string, boolean

a = 7 + 3j # complex
print(str(a)) # '7+3j'
print(bool(a)) # True

print(int(a)) # TypeError: can't convert complex to int
print(float(a)) # TypeError: can't convert complex to float
print(list(a)) # TypeError: 'complex' object is not iterable
print(tuple(a)) # TypeError: 'complex' object is not iterable
print(set(a)) # TypeError: 'complex' object is not iterable
print(dict(a)) # TypeError: 'complex' object is not iterable
```

```python
# bool datatype - can be converted into int, float, complex, string

a = True # bool
print(int(a)) # 1
print(float(a)) # 1.0
print(complex(a)) # 1+0j
print(str(a)) # 'True'

print(list(a)) # TypeError: 'bool' object is not iterable
print(tuple(a)) # TypeError: 'bool' object is not iterable
print(set(a)) # TypeError: 'bool' object is not iterable
print(dict(a)) # TypeError: 'bool' object is not iterable
```

```python
# string datatype - can be converted into int, float, complex, boolean

string = 'python' # string
print(list(string)) # ['p', 'y', 't', 'h', 'o', 'n']
print(tuple(string)) # ('p', 'y', 't', 'h', 'o', 'n')
print(set(string)) # {'p', 'y', 't', 'h', 'o', 'n'}
print(bool(string)) # True

print(int(string)) # ValueError: invalid literal for int() with base 10: 'python'
print(float(string)) # ValueError: could not convert string to float: 'python'
print(complex(string)) # ValueError: complex() arg is a malformed string
print(dict(string)) # TypeError: cannot convert dictionary update sequence element #0 to a sequence
```

```python
# list datatype - can be converted into string, boolean, set, tuple

list1 = [1, 2, 3, 4, 5] # list
print(str(list1)) # '[1, 2, 3, 4, 5]'
print(bool(list1)) # True
print(set(list1)) # {1, 2, 3, 4, 5}
print(tuple(list1)) # (1, 2, 3, 4, 5)

print(int(list1)) # TypeError: int() argument must be a string, a bytes-like object or a number, not 'list'
print(float(list1)) # TypeError: float() argument must be a string or a number, not 'list'
print(complex(list1)) # TypeError: complex() first argument must be a string order a number, not 'list'
print(dict(list1)) # TypeError: cannot convert dictionary update sequence element #0 to abstract a sequence 
```

```python
# tuple datatype - can be converted into string, boolean, set, list

tuple1 = (1, 2, 3, 4, 5) # tuple
print(str(tuple1)) # '(1, 2, 3, 4, 5)'
print(bool(tuple1)) # True
print(set(tuple1)) # {1, 2, 3, 4, 5}
print(list(tuple1)) # [1, 2, 3, 4, 5]

print(int(tuple1)) # TypeError: int() argument must be a string, a bytes-like object or a number, not 'tuple'
print(float(tuple1)) # TypeError: float() argument must be a string or a number, not 'tuple'
print(complex(tuple1)) # TypeError: complex() first argument must be a string order a number, not 'tuple'
print(dict(tuple1)) # TypeError: cannot convert dictionary update sequence element #0 to abstract a sequence    
```

```python
# set datatype - can be converted into string, boolean, list, tuple

set1 = {1, 2, 3, 4, 5} # set
print(str(set1)) # '{1, 2, 3, 4, 5}'
print(bool(set1)) # True
print(list(set1)) # [1, 2, 3, 4, 5]
print(tuple(set1)) # (1, 2, 3, 4, 5)

print(int(set1)) # TypeError: int() argument must be a string, a bytes-like object or a number, not 'set'
print(float(set1)) # TypeError: float() argument must be a string or a number, not 'set'
print(complex(set1)) # TypeError: complex() first argument must be a string order a number, not 'set'
print(dict(set1)) # TypeError: cannot convert dictionary update sequence element #0 to abstract a sequence  

```

```python
# dictionary datatype - can be converted into string, boolean, list, tuple, set

dict1 = {'a': 1, 'b': 2, 'c': 3} # dictionary
print(str(dict1)) # "{'a': 1, 'b': 2, 'c': 3}"
print(bool(dict1)) # True   
print(list(dict1)) # ['a', 'b', 'c']
print(tuple(dict1)) # ('a', 'b', 'c')
print(set(dict1)) # {'a', 'b', 'c'}

print(int(dict1)) # TypeError: int() argument must be a string, a bytes-like object or a number, not 'dict'
print(float(dict1)) # TypeError: float() argument must be a string or a number, not 'dict'
print(complex(dict1)) # TypeError: complex() first argument must be a string order a number, not 'dict'
```
