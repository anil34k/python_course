```python
jumping statemnts

is used to change the flow of execution in looping statements

1. break statement - it is used to exit from the loop when a certain condition is met
2. continue statement - it is used to skip the current iteration of the loop and move to the next iteration
3. pass statement - it is used as a placeholder when a statement is required syntactically but no action is needed
```

```python
# break statement

# wap to print numbers from 1 to 4 using 1 to 20 range and break statement

for i in range(1, 21):
    if i == 5:
        break
    print(i) # 1 2 3 4

print('\n')

# wap to print character upto e using a to z range and break statement

for j in 'abcdefghi':
    if j == 'f':
        break
    print(j) # a b c d e

print('\n')

# or

for i in range(ord('a'), ord('z') + 1):
    if chr(i) == 'f':
        break
    print(chr(i)) # a b c d e

# wap to stop the loop when the number divisible by 7 in 1 to 20 range using break statement
for i in range(1, 21):
    if i % 7 == 0:
        break
    print(i) # 1 2 3 4 5 6
    
```
