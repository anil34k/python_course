```python
def greet(name):
    print(f"Hello {name}")

# 1-perform a task
# 2-return a value
def greeting1(name):
    return f"Hello {name}"

message = greeting1("Anil")
print(message)

def fizz_buzz(input):
  if input % 3 == 0:
    return 'fizz'
  if input % 5 == 0:
    return 'buzz'
  if (input % 3 == 0) and (input % 5 == 0):
    return 'fizzbuzz'
  return input

print(fizz_buzz(15)) 
```
