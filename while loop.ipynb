```python
# wap to check the given number is prime or not
# prime number means a number greater than 1,that has only 2 divisors (1 and itself) 
num = int(input())
i = 1
count = 0
while num >= i:
    if num % i == 0:
        count += 1
    i += 1
if count == 2:
    print('prime')
else :
    print('not prime')
```

```python
# gievn number armstrong number or not
# armstriong numbers means a number that is equal to the sum of its own digits and rised to the power of number of digits 
# example 153 -> 1**3 + 5**3 + 3**3

num = 153
temp = num
sum = 0
while num > 0 :
    digit = num % 10
    sum += digit ** len(str(num))
    num = num // 10
if temp == sum:
    print('armstrong')
else:
    print('not armstrong')
```

```python
# print fibonacci series
# example 0 1 1 2 3 5 8 13
n = int(input())
a, b = 0, 1
count = 0  # Initialize count correctly

# Run the loop until we print 'n' numbers
while count < n:
    # Print the current number
    print(a, end=" ")
    
    # Calculate the next numbers in the series
    a, b = b, a + b
    
    # Increment the counter
    count = count + 1
```

```python
# to convert lower case string to uppercase
s = 'python'
i = 0
result = ''
while i < len(s):
    if s[i] >= 'a' and s[i] <= 'z':
        result = result + chr(ord(s[i]) - 32)
    else:
        result = result + s[i]
    i += 1
print(result)
```

```python
# to toggle characters case

s = 'PYTHON'
i = 0
result = ''
while i < len(s):
    ch = s[i]
    if ch >= 'A' and ch <= 'Z':
        result = result + chr(ord(ch)+32)
    elif ch >= 'a' and ch <= 'z':
        result = result + chr(ord(ch)-32)
    else :
        result = result + ch
    i = i + 1
print(result)
```

```python
# wap to perform atm pin authentication and check balance, deposit, withdraw and exit 
# operation using while loop

pin = '1234'
balance = 1000.0
attempts = 0

while attempts < 3:
    entered = input('enter the pin')
    if entered == pin:
        break
    else:
        print('wrong pin')
        attempts += 1
else :
    print('account locked')

# main logic using while loop
if attempts < 3:
    while True:
        print('1. check balance')
        print('2. deposit')
        print('3. withdraw')
        print('4. exit')
        
        choice = input('enter your choice')
        
        if choice == '1':
            print('your balance is:', balance)
        elif choice == '2':
            amount = float(input('enter amount to deposit'))
            balance += amount
            print('deposit successful, new balance is:', balance)
        elif choice == '3':
            amount = float(input('enter amount to withdraw'))
            if amount <= balance:
                balance -= amount
                print('withdraw successful, new balance is:', balance)
            else:
                print('insufficient funds')
        elif choice == '4':
            print('thank you for using atm')
            break
        else:
            print('invalid choice')
```
