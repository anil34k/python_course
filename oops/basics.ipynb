```python
# (7) WAP to create function inside class & to access class variables using object
class student:
    name = 'martin'
    age = 26

    def display(self):
        print(f'student name: {self.name}')
        print(f'student age: {self.age}')

s1 = student()
s1.display()

# or without using self
class students:
    name = 'martin'
    age = 26

    def display():
        print(f'student name: {students.name}')
        print(f'student age : {students.age}')

students.display()
```

```python
# (9) Example using function as objects
class employee:
    dname = 'research'

    def details(self, ename, sal):
        self.ename = ename
        self.sal = sal

    def display(self):
        print(f'employee name : {self.ename}')
        print(f'department name : {self.dname}')
        print(f'salary is : {self.sal}')
e1 = employee()
e1.details('scott', 45000)
e1.display()

e2 = employee()
e2.details('smith', 95000)
e2.display()
```

```python
# 10) WAP to perform checkbalance, deposit & withdraw operation
class bank:
    bankname = 'icici'      # class variables

    def details(self, cname, balance):     # instance variables
        self.cname = cname
        self.balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.balance = self.balance + amount
            print(f'{amount} has been deposited')
            print(f'after deposit updated balance : {self.balance}')
        else:
            print('amount should be greater than zero')
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance = self.balance - amount
            print(f'{amount} has been debited')
            print(f'after withdraw updated balance : {self.balance}')
        else:
            rint('Insufficient balance')
    def display(self):
        print(f'customer name: {self.cname}')
        print(f'initial bank balance is : {self.balance}')
c1 = bank()
c1.details('smith', 9800)
c1.display()
c1.deposit(1000)
c1.withdraw(500)
c2 = bank()
c2.details('scott', 50000)
c2.display()
c2.deposit(1000)
c2.withdraw(2000)
```
